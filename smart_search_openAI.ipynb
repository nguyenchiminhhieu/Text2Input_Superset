### Hàm này để đọc file Excel thành một DF

```python

def read_excel_to_df(file_path: str) -> pd.DataFrame:
    
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File not found: {file_path}")
    
    
    if not file_path.endswith(('.xlsx', '.xls')):
        raise ValueError("File must be an Excel file (.xlsx or .xls)")
    
    try:
        
        df = pd.read_excel(file_path)
        
       
        if df.empty:
            print("Warning: Excel file is empty")
            
        return df
    
    except Exception as e:
        raise Exception(f"Error reading Excel file: {str(e)}")
```

### Hàm này để gen ra 3 tham số "groupby", "viz_type", "slice_name" từ yêu cầu của người dùng
 Em sử dụng Llama 3.2

```python
def get_groupby_and_chart_type_from_request(user_request, columns_str):
    llm = ChatOllama(model="llama3.2", temperature=0)
    
    prompt = f"""
    At the user's request, specify which column should be used for the "groupby" operation, the type of chart that the user is requesting (if not required, automatically select the most appropriate type), and the name of the chart (slice_name). There are 4 types of charts: pie_charts(biểu đồ tròn), bar_charts(biểu đồ cột), line_charts( biểu đồ đường), and number_charts(biểu đồ số).
    
    Available columns for groupby: {columns_str}
    
    User request: "{user_request}"
    
    Please return only one JSON string with the following structure, Don't add anything else including explanation:
    {{
      "groupby": ["column1", "column2"],
      "viz_type": "chart_type",
      "slice_name": "name_of_the_chart"
    }}
    Where:
    - "groupby" is a list of columns that should be used for the "groupby" operation based on the user's request.
    - "viz_type" is a type of chart for analysis based on user requirements.t.
    - "slice_name" is a descriptive name for the chart based on the analysis.
    """

    
    response = llm.invoke(prompt)  

    
    response_content = response.content  
    
    return response_content
   
```

### Đây là hàm để tạo thành một API hoàn chỉnh

```python
def generate_chart_from_template(output):
    
    viz_type = output.get("viz_type")
    groupby_or_axis = output.get("groupby")  
    slice_name = output.get("slice_name")
    
    if viz_type not in TEMPLATES:
        raise ValueError(f"Loại biểu đồ '{viz_type}' không cos.")
    
    template = TEMPLATES[viz_type]
    
    
    try:
        if isinstance(groupby_or_axis, list):
            groupby_or_axis = [item.encode('utf-8').decode('utf-8') if isinstance(item, str) else item 
                             for item in groupby_or_axis]
        
        if viz_type == "pie_charts":
           
            template["params"] = json.loads(template["params"]) if isinstance(template["params"], str) else template["params"]
            template["params"]["groupby"] = groupby_or_axis
            template["params"] = json.dumps(template["params"], ensure_ascii=False)
            
        elif viz_type in ["bar_charts", "line_charts"]:
            
            template["params"] = json.loads(template["params"]) if isinstance(template["params"], str) else template["params"]
            template["params"]["x_axis"] = groupby_or_axis
            template["params"] = json.dumps(template["params"], ensure_ascii=False)
            
        elif viz_type == "number_charts":
            # Em đang đợi Định a ạ. EM chưa có template với cả chưa biết xử lý logic như nào
            pass
            
        template["slice_name"] = slice_name
        return template

    except Exception as e:
        raise ValueError(f"Lỗi khi tạo template biểu đồ: {e}")
```

### Đây là hàm cuối cùng với input là file Excel và yêu cầu từ người dùng ---> nó sẽ trả về API phù hơp

```python
def process_user_request_with_file(file_path: str, user_request: str):

    try:
        df = read_excel_to_df(file_path)
        
        if df.empty:
            raise ValueError("File Excel không chứa dữ liệu.")
        
        columns = df.columns.tolist()
        columns_str = ', '.join([f'"{col}"' for col in columns])
        
        output_str = get_groupby_and_chart_type_from_request(user_request, columns_str)

        try:
            output = json.loads(output_str)
        except json.JSONDecodeError as e:
            raise ValueError(f"Lỗi khi chuyển đổi JSON output: {e}")
        
        chart = generate_chart_from_template(output)
        
        return chart

    except Exception as e:
        raise ValueError(f"Lỗi khi xử lý yêu cầu với file Excel: {e}")

if __name__ == "__main__":
    file_path = "./data_example.xlsx" 
    user_request = "Tạo biểu đồ tròn thể hiên tỷ lệ tôn giáo"
    
    try:
        chart = process_user_request_with_file(file_path, user_request)
        print(chart)
    except ValueError as e:
        print(f"Lỗi: {e}")
```
