### Hàm này để đọc file Excel thành một DF

````python

def read_excel_to_df(file_path: str) -> pd.DataFrame:
    
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File not found: {file_path}")
    
    
    if not file_path.endswith(('.xlsx', '.xls')):
        raise ValueError("File must be an Excel file (.xlsx or .xls)")
    
    try:
        
        df = pd.read_excel(file_path)
        
       
        if df.empty:
            print("Warning: Excel file is empty")
            
        return df
    
    except Exception as e:
        raise Exception(f"Error reading Excel file: {str(e)}")
````

### Hàm này để gen ra 3 tham số "groupby", "viz_type", "slice_name" từ yêu cầu của người dùng
 Em sử dụng OpenAI

````python
import openai

openai.api_key = "xxx"

def get_groupby_and_chart_type_from_request(user_request, columns_str):
  
    prompt = f"""
    At the user's request, determine the following:
    1. The column(s) to be used for the "groupby" operation.
    2. The type of chart the user is requesting (if not specified, select the most appropriate type automatically).
    3. A descriptive name for the chart (slice_name).
    4. If the chart type is "number_charts", provide:
      - **Aggregate**: The aggregation method, limited to COUNT or COUNT_DISTINCT.
      - **Label**: The specific aggregation expressed in SQL-style, e.g., COUNT(column_name) or COUNT_DISTINCT(column_name).

    Available chart types:
    - "pie_charts" (biểu đồ tròn)
    - "bar_charts" (biểu đồ cột)
    - "line_charts" (biểu đồ đường)
    - "number_charts" (biểu đồ số)

    A list of available columns for the groupby operation: {columns_str}

    User request: "{user_request}"

    Please return a single JSON string with the following structure (do not include any additional text or explanation):
    {{
      "groupby": ["column1", "column2"],
      "viz_type": "chart_type",
      "slice_name": "name_of_the_chart",
      "aggregate": "aggregation_method",
      "label": "SQL-style_aggregation"
    }}
    Where:
    - "groupby": A list of columns selected for the "groupby" operation, based on the user's request.
    - "viz_type": The type of chart selected for analysis, based on user requirements or the most suitable type.
    - "slice_name": A descriptive name for the chart that reflects the analysis or data visualization purpose.
    - "aggregate": (Only for number_charts) The aggregation method for the data, such as COUNT, SUM, etc.
    - "label": (Only for number_charts) The SQL-style aggregation expression, such as `COUNT(column_name)` or `SUM(column_name)`.
    """

    response = openai.ChatCompletion.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are a helpful assistant that responds with structured JSON outputs."},
            {"role": "user", "content": prompt}
        ],
        temperature=0
    )

    response_content = response["choices"][0]["message"]["content"]
   
    
    
    return response_content
````

### Đây là hàm để tạo thành một API hoàn chỉnh

````python
def generate_chart_from_template(output):
    viz_type = output.get("viz_type")
    groupby_or_axis = output.get("groupby")
    aggregate = output.get("aggregate")
    label = output.get("label")
    slice_name = output.get("slice_name")
    
    if viz_type not in TEMPLATES:
        raise ValueError(f"Loại biểu đồ '{viz_type}' không cos.")
    
    template = TEMPLATES[viz_type]
    
    
    try:
        if isinstance(groupby_or_axis, list):
            groupby_or_axis = [item.encode('utf-8').decode('utf-8') if isinstance(item, str) else item 
                             for item in groupby_or_axis]
        
        if viz_type == "pie_charts":
           
            template["params"] = json.loads(template["params"]) if isinstance(template["params"], str) else template["params"]
            template["params"]["groupby"] = groupby_or_axis
            template["params"] = json.dumps(template["params"], ensure_ascii=False)
            
        elif viz_type in ["bar_charts", "line_charts"]:
            
            template["params"] = json.loads(template["params"]) if isinstance(template["params"], str) else template["params"]
            template["params"]["x_axis"] = groupby_or_axis
            template["params"] = json.dumps(template["params"], ensure_ascii=False)
            
        elif viz_type == "number_charts":
            template["params"] = json.loads(template["params"]) if isinstance(template["params"], str) else template["params"]
            template["params"]["metric"]["label"] = label
            template["params"]["metric"]["column"]["column_name"] = groupby_or_axis
            template["params"]["metric"]["aggregate"] = aggregate
            template["params"] = json.dumps(template["params"], ensure_ascii=False)

            pass
            
        template["slice_name"] = slice_name
        return template

    except Exception as e:
        raise ValueError(f"Lỗi khi tạo template biểu đồ: {e}")
````

### Đây là hàm cuối cùng với input là file Excel và yêu cầu từ người dùng ---> nó sẽ trả về API phù hơp

````python
def process_user_request_with_file(file_path: str, user_request: str):

    try:
            df = read_excel_to_df(file_path)

            if df.empty:
                raise ValueError("File Excel không chứa dữ liệu.")

            columns = df.columns.tolist()
            columns_str = ', '.join([f'"{col}"' for col in columns])

            output_str = get_groupby_and_chart_type_from_request(user_request, columns_str)

            output_str = output_str.strip().lstrip('```').lstrip('json').rstrip('```').strip()

            

            if not output_str:
                raise ValueError("Không nhận được JSON hợp lệ từ OpenAI.")

            output = json.loads(output_str)
            chart = generate_chart_from_template(output)

            return chart

    except Exception as e:
            raise ValueError(f"Lỗi khi xử lý yêu cầu với file Excel: {e}")

if __name__ == "__main__":
    file_path = "./data_example.xlsx" 
    user_request = "Tạo biểu đồ số thể hiên điểm GPA trung bình của sinh viên"
    
    try:
        chart = process_user_request_with_file(file_path, user_request)
        print(chart)
    except ValueError as e:
        print(f"Lỗi: {e}")
````
